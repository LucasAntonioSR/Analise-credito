**Abaixo será feito a arvore de decisão para saber qual situação em que é mais propício de haver default.**

```python
df_used_tree=df[['default','idade','dependentes','meses_de_relacionamento','qtd_produtos','iteracoes_12m','meses_inativo_12m','limite_credito','valor_transacoes_12m','education','relacionamento','cartao']]
df_used_tree
```

```python
y = df_used_tree['default']
```

```python
x = df_used_tree.drop('default', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=100)
```

```python
clf = DecisionTreeClassifier(random_state=100)
clf = clf.fit(x_train, y_train)
```

```python
plt.figure(figsize=(25, 20))
plot_tree(clf,
         filled= True,
         fontsize= 10,
         class_names=['Maior chance', 'Menor chance'],
         feature_names=x.columns);
```

```python
display = ConfusionMatrixDisplay.from_estimator(clf, x_train, y_train,
                                               display_labels=['Maior chance', 'Menor chance'])
display
```

```python
predict = clf.predict(x_train)
train_accuracy = accuracy_score(y_train, predict)
train_accuracy
```

```python
clf_predict = clf.predict(x_test)
clf_predict
```

```python
display = ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test,
                                               display_labels=['Maior chance', 'Menor chance'])
display
```

```python
test_ = clf.predict(x_test)
test_accuracy = accuracy_score(y_test, test_)
test_accuracy
```
